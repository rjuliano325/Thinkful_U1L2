```python
# Choose one variable and plot that variable four different ways.

# Load the desired dataset in to a DataFrame
diamonds = sns.load_dataset('diamonds')

# Set figure size 
plt.figure(figsize=(10,7))

# First subplot
plt.subplot(2,2,1)
sns.distplot(diamonds['carat'], kde=False, color='g', bins=12)
plt.title('Non Normed Histogram')

# Second subplot
plt.subplot(2,2,2)
sns.rugplot(diamonds['carat'].head(1000)) # Limit to first 1,000 observations for speed
plt.title('RUG Plot')

# Third Subplot
plt.subplot(2,2,3)
sns.kdeplot(diamonds['carat'], color='k',shade=True)
plt.title('KDE Plot')

# Fourth subplot
plt.subplot(2,2,4)
sns.distplot(diamonds['carat'], norm_hist=True)
plt.title('Normed Histogram + KDE Plot')
             
# Remove extraneous axis ticks, format for better layout, and show plots    
sns.despine()
plt.tight_layout()
plt.show()

print('Histogram: Good for seeing the distribution of the data. Con is you can only see univariate data, and shape is sensitive to number of bins used.')
print('RUG Plot: Good for seeing the the density of a distribution, but does not convey numbers of occurences like histogram.')
print('KDE Plot: Shows the probability density of data, also has the advantage that it can show bivariate data.')
print('Normed Hist + KDE Plots: You can see the distribution of the data and its probability distribution.')
```

```python
# Choose two continuous variables, and plot them three different ways.

# Look at the relationship between carat and price

sns.regplot(x='carat', y='price', data=diamonds, fit_reg=False, color='r')
plt.title('Carat versus Price')

plt.show()

print("LM Plot: Show relationships between two variables. Can get cluttered and difficult to see a relationship without a regression line or transparency.")
```

```python
# Look at the relationship between carat and price

sns.jointplot(x='carat', y='price', data=diamonds,kind='hex', size=6, color='purple')
plt.title('Carat versus Price')

plt.show()

print('Hex Plot: Shows relationship between two variables, includes histograms to see data distribution.')
print('This plot may not be ideal if you want to see all data points.')
```

```python
# Look at the relationship between carat and price

sns.lmplot(x='carat', y='price', data=diamonds, fit_reg=True, hue='clarity')
plt.title('Carat versus Price')

plt.show()

print('Scatter Plot with Regressions: Shares same benefits of Scatter plot mentioned previously.')
print('Has the additional benefit of regression lines by category, to more easily see the relationships between variables.')
```

```python
#Choose one continuous variable and one categorical variable, and plot them six different ways.
plt.figure(figsize=(10,7))
plt.title('Title')

#stripplot
plt.subplot(2,3,1)
plt.title('Price By Color')
sns.stripplot(x='color', y='price', data=diamonds, jitter=True)

#boxplot
plt.subplot(2,3,2)
plt.title('Price By Color')
sns.boxplot(x='color', y='price', data=diamonds,)

#barplot
plt.subplot(2,3,3)
plt.title('Price By Color')
sns.barplot(x='color', y='price', data=diamonds,)

#countplot
plt.subplot(2,3,4)
plt.title('Price By Color')
sns.countplot(x='color', data=diamonds,)

#violinplot
plt.subplot(2,3,5)
plt.title('Price By Color')
sns.violinplot(x='color', y='price', data=diamonds,)

#pointplot
plt.subplot(2,3,6)
plt.title('Price By Color')
sns.pointplot(x='color', y='price', data=diamonds,)

#visual enhancement and 'show'
sns.despine()
plt.tight_layout()
plt.show()

print('Strip Plot: Good for smaller observations of data; is difficult to see where the majority of points fall.')
print('Box Plot: Good for seeing the quartiles and outliers. For data with many outliers, may not be ideal.')
print('Bar Plot: Shows the central tendency, mean in this case, of prices by color with confidence intervals.')
print('Count Plot: Good for understanding the numbers of observations a varible in a dataset.')
print('Violin Plot: Useful for showing quartiles as well as distribution. Not the best choice for seeing granular data.')
print('Point Plot: Useful for seeing confidence interval and value by category. Similar to Bar Plot without filled columns.')
```
